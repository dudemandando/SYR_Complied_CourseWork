```python
#import the nltk
import nltk

#Downlad the Gutenburg
nltk.download('gutenberg', quiet=True)

# Check the files, titles of books 
nltk.corpus.gutenberg.fileids()
```

# Title Selection

Here I've chosen to utilize two texts from Gutenberg data provided within the "nltk".  

*   Hamlet by Shakespeare
*   Macbeth by Shakespeare



```python
# Extract the selected titles
print("Selecting: ", nltk.corpus.gutenberg.fileids()[15])
print("Selecting: ", nltk.corpus.gutenberg.fileids()[16])

hamlet_title = nltk.corpus.gutenberg.fileids()[15]
macbeth_title = nltk.corpus.gutenberg.fileids()[16]

hamlet_text = nltk.corpus.gutenberg.raw(hamlet_title)
macbeth_text = nltk.corpus.gutenberg.raw(macbeth_title)
```

# Tokenize the Texts


```python
# Required for tokenization
nltk.download('punkt', quiet=True)
```

```python
hamlet_tokens = nltk.word_tokenize(hamlet_text)
macbeth_tokens = nltk.word_tokenize(macbeth_text)

hamlet_words = [w.lower() for w in hamlet_tokens]
macbeth_words =  [w.lower() for w in macbeth_tokens]
```

# Stop Words

```python
# Download the Nltk's stop words
nltk.download('stopwords', quiet=True)
```

```python
#Store the stop words
nltk_stopwords = nltk.corpus.stopwords.words('english')
print("Total Stop Words: ", len(nltk_stopwords))
```

#  Modifying the Stop Words List

Being that I've selected two of Shakespeare's works (written in a older style of english) I've needed to modify the stop words collection. Not only are these selected works written in an older form of english, they are also playwrights which will contain characters a punctuation which varies from that of a more common group of text (novels, textbooks, blog posts etc...).

The old english combined with playwright styling has caused me to include (below) common punctuation and old english words contained with the variable "stopwords" (I've neglected to list here for readability sake).

**Important Note:**

It is important to note that I started analysis of these texts knowing beforehand that they are playwrights written approximately 400 years ago. Had I not had this knowledge prior, I would have not had the immediate intitial intution to expand the stopwords list. At the same time, it is situations like this, where we find multiple bigrams to include single punctuation characters should be a signal to users of the structure and format of the overall text. 


```python
# Adding additional Stop Words, common punctuation chars
stopwords = nltk_stopwords + [',','.',':',';', "'s",'?', 'thou', 'thy',"'",'would', 'well','the','th', '[',']',"'t", 'enter']

#Remove those stop words from the selected texts
hamlet_words_stop = [w for w in hamlet_words if not w in stopwords]
macbeth_words_stop =[w for w in macbeth_words if not w in stopwords]
```

# List the top 50 words by frequency (normalized by the length of the document)

```python
from nltk import FreqDist
n_dist_hamlet = FreqDist(hamlet_words_stop)
n_dist_macbeth = FreqDist(macbeth_words_stop)

#List the top 50 Words by Frequency via the most common method
print(n_dist_hamlet.most_common(50))
print(n_dist_macbeth.most_common(50))
```

# List the top 50 bigrams by frequencies

```python
#Use a Generator for Performance
bigram_gen_macbeth = nltk.bigrams(macbeth_words_stop)
bigram_gen_hamlet = nltk.bigrams(hamlet_words_stop)

```

```python
print("Top 50 bigrams: Hamlet")
for x in range(50):
  print(x ,":", next(bigram_gen_hamlet))
```

```python
print("Top 50 bigrams: Macbeth")
for x in range(50):
  print(x ,":", next(bigram_gen_macbeth))
```

# List the top 50 bigrams by their Mutual Information scores (using min frequency 5)

```python
from nltk import collocations
# Call the bigram method within NLTK
bigram_measures = collocations.BigramAssocMeasures()

```

```python

# Apply the Macbeth and Hamlet Text to the bigram method
macbeth_finder = collocations.BigramCollocationFinder.from_words(macbeth_words_stop)

hamlet_finder = collocations.BigramCollocationFinder.from_words(hamlet_words_stop)

#Apply a frequency filter with value of '5'
macbeth_finder.apply_freq_filter(5)
hamlet_finder.apply_freq_filter(5)
```

```python
macbeth_scores = macbeth_finder.score_ngrams(bigram_measures.pmi)
hamlet_scores = hamlet_finder.score_ngrams(bigram_measures.pmi)
len(macbeth_scores)
```

# Question: Can the Use of Trigrams Prove To Be More Contextually Insightful with Shakespeare's Hamlet and Macbeth?

```python
from nltk import trigrams
trigram_measures = nltk.collocations.TrigramAssocMeasures()
```

```python
# Reset the stopwords 
#Store the stop words
nltk_stopwords = nltk.corpus.stopwords.words('english')
print("Total Stop Words: ", len(nltk_stopwords))

# Adding additional Stop Words, common punctuation chars
stopwords = nltk_stopwords + [',','.',':',';', "'s",'?', 'thou', 'thy',"'",'would', 'well','the','th', '[',']',"'t", 'enter', 'exeunt', 'good', 'lord']

#Remove those stop words from the selected texts
hamlet_words_stop = [w for w in hamlet_words if not w in stopwords]
macbeth_words_stop =[w for w in macbeth_words if not w in stopwords]


#examine frequency distribution with reduced stop words
n_dist_hamlet = FreqDist(hamlet_words_stop)
n_dist_macbeth = FreqDist(macbeth_words_stop)

#List the top 50 Words by Frequency via the most common method
print(n_dist_hamlet.most_common(50))
print(n_dist_macbeth.most_common(50))
```

```python
macbeth_finder_tri = collocations.TrigramCollocationFinder.from_words(macbeth_words_stop)
hamlet_finder_tri = collocations.TrigramCollocationFinder.from_words(hamlet_words_stop)
```

```python
macbeth_tri_scored = macbeth_finder_tri.score_ngrams(trigram_measures.raw_freq)
hamlet_tri_scored = hamlet_finder_tri.score_ngrams(trigram_measures.raw_freq)
```
