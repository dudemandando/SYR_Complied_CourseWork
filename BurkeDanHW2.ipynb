```python
#Import Required Dependencies
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from google.colab import files
import io
import re

uploaded = files.upload
```

```python
print("Original Source of Data is https://satoshi.nakamotoinstitute.org/emails/cryptography/3/#selection-55.0-75.16")

#Create Path Strings to import data
emailOnePath = "/content/SatoshiEmail_1.txt"
emailTwoPath = "/content/SatoshiEmail_2.txt"
emailThreePath = "/content/SatoshiEmail_3.txt"
```

```python
def ReadReturnTxtFile(path):
  fileData = open(path, "r")
  fileDataContent = fileData.read()
  return fileDataContent
```

```python
#Import Data
emailOneRaw = ReadReturnTxtFile(emailOnePath)
emailTwoRaw = ReadReturnTxtFile(emailTwoPath)
emailThreeRaw = ReadReturnTxtFile(emailThreePath)
```

```python
emailOneRaw
```

```python
#Create Method to remove special characters from each raw string
def RemoveSpecialChar(specialChar, strToUse):
  cleanString = re.sub('\n',' ', strToUse )
  return cleanString

#Create List of all Emails
emailList = [emailOneRaw, emailTwoRaw, emailThreeRaw]

emailList[0] = RemoveSpecialChar('\n', emailList[0])
emailList[1] = RemoveSpecialChar('\n', emailList[1])
emailList[2] = RemoveSpecialChar('\n', emailList[2])
```

```python
#Instantiate Count Vectorizer
satoshiCV = CountVectorizer(input = "content", stop_words=None)

#Create Matrix of all data
satoshiMat = satoshiCV.fit_transform(emailList)

#Verify Type
print(type(satoshiMat))

#Retrieve Column Names
satoshiColNames = satoshiCV.get_feature_names_out()
print("Column Names Length:",len(satoshiColNames))
satoshiColNames
```

```python
satoshipDF = pd.DataFrame(satoshiMat.toarray(), columns=satoshiColNames)
```
